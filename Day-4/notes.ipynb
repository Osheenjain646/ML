## Sets in Python

## Not follow order and no duplicates

```python
## create set
my_set=(1,2,2,5)
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,3,5,565,5,3,6,5])
print(type(my_empty_set))
print(my_set)
```

```python
## basic operations
## adding and removing elements
my_set.add(7)
print(my_set)
```

```python
my_set.add(7)
print(my_set)
```

```python
## removing elements
my_set.remove(7)
print(my_set)
```

```python
my_set.remove(254)
```

```python
## removing elements safely
my_set.discard(254)
print(my_set)
```

```python
## pop method
popped_element=my_set.pop() ## pop out an arbitrary element
print("popped element:",popped_element)
print("set after pop:",my_set)
```

```python
## clear all elements
my_set.clear()
print(my_set)
```

```python
## set membership test
my_set={1,2,3,4,5}
print(3 in my_set)  ## give boolean output if element is present true/false
print(10 in my_set)
```

```python
## mathematical set operations
set1={1,2,3,4,5,5,6}
set2={4,5,6,7,8,9}

## union
union_set=set1.union(set2)  ## add all unique elements from both sets
print("union:",union_set)
```

```python
## intersection
intersection_set=set1.intersection(set2)
print(intersection_set)  ## common elements in both sets
```

```python
set1.intersection_update(set2)
print(set1)  ## set1 now contains only elements also in set2 updated with common elenments
```

```python
set1={1,2,3,4,5,5,6}
set2={4,5,6,7,8,9}
## difference
print(set1.difference(set2))  ## elements in set1 but not in set2
print(set2.difference(set1))  ## elements in set2 but not in set1

set1.difference_update(set2)
print(set1)  ## set1 updated to contain only elements not in set2
```

```python
## symmetric difference
set1={1,2,3,4,5,5,6}
set2={4,5,6,7,8,9}
set1.symmetric_difference(set2)  ## elements in either set1 or set2 but not in both will be returned
```

```python
## set methods
set1={1,2,3,4,5}
set2={4,5,6,7,8}
## is subset
print(set1.issubset(set2))  ## check if set1 is subset of set2
print(set2.issuperset(set1))  ## check if set2 is superset of set1

## superset
print(set2.issuperset(set1))  ## check if set2 is superset of set1 have all elements of set1
print(set1.issubset(set2))  ## check if set1 is subset of set2 have all elements in set2
## is disjoint
## check if two sets have no elements in common
```

```python
lst=[1,2,2,3,4,4,5]
print(set(lst)) ## remove duplicates from list using set
```

```python
## counting all unique elements in a text
text="hello world welcome to the world of python programming hello hello"
unique_words=set(text.split())
print(unique_words)
print(len(unique_words))
```

## Dictionaries

```python
## unordered collection of unique elements , key value pair , immutable
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(type(empty_dict))
print(empty_dict)
```

```python
## key value pair
student={"name":"Osheen", "age":20, "course":"ML"}
print(student)
print(type(student))
```

```python
## error
student={"name":"Osheen", "age":20, "course":"ML","name":45}
print(student)  ## keys are unique , last assignment will be retained or overwritten
```

```python
## accessing dictionary elements
student={"name":"Osheen", "age":20, "course":"ML"}
print(student['name'])
print(student['age'])
print(student['course'])

## default method get
print(student.get('age'))
print(student.get('grade'))  ## returns None if key not found
print(student.get('name'))

print(student.get('grade','Not Available'))  ## returns 'Not Found' if key not found as default value provided
```

```python
## modifying dictionary elements
## dictionary is mutable, you can change its content as add update or delete elements
print(student)
```

```python
student['age']=21 ## update age
print(student)
student["address"]="India" ## add new key value pair
print(student)
```

```python
## deleting key
del student['course'] ## delete key and value pair
print(student)
```

```python
## dictionary methods

keys=student.keys()
print(keys)
values=student.values()
print(values)
```

```python
## to get key value pairs
items=student.items()
print(items) ## giving list of tuples of key value pairs
```

```python
## shallow copy of dictionary
student_copy=student
print(student_copy)
print(student)
```

```python
student["name"]="New Name"  ## update name in original dictionary and also in copy as both refer to same object
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()  ## create a shallow copy of dictionary
print(student_copy1)
print(student)
```

```python
student["name"]="New Name2" ## update name in original dictionary
print(student)
print(student_copy1)  ## no change in copy
```

```python
## iterating through dictionary

## using for loop to iterate dictionary keys , values and items

## iterate keys
for key in student.keys():
    print(key)
    
```

```python
## over values
for value in student.values():
    print(value)
```

```python
## iterating through items
for key,value in student.items():
    print(f"{key} : {value}")
```

```python
## NESTED DICTIONARIES
students={
    "student1":{"name":"Osheen","age":20},
    "student2":{"name":"Alice","age":22}
}
print(students)
```

```python
## accessing nested dictionary elements
print(students["student1"]["name"])
print(students["student1"]["age"])
```

```python
students.items()
```

```python
## iterating through nested dictionary
for student_id,student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key,value in student_info.items():
        print(f"   {key} : {value}")
        
```

```python
## dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dictionary comprehension

even_squares={x:x**2 for x in range(10) if x%2==0}
print(even_squares)
```

```python
## practical examples

## use a dictinary to count frequency of elements in a list
numbers=[1,2,2,3,4,4,5,1,2,3,4,5,5,5]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merging dictionaries

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
