## 1Ô∏è‚É£ Count characters ignoring spaces
def char_count(s):
    counts = {}
    for ch in s:
        if ch != " ":          # ignore spaces
            counts[ch] = counts.get(ch, 0) + 1
    return counts

print(char_count("hello world"))


Output:

{'h': 1, 'e': 1, 'l': 3, 'o': 2, 'w': 1, 'r': 1, 'd': 1}

2Ô∏è‚É£ Case-insensitive character count
def char_count(s):
    s = s.lower()              # convert to lowercase
    counts = {}
    for ch in s:
        counts[ch] = counts.get(ch, 0) + 1
    return counts

print(char_count("Python"))


Output:

{'p': 1, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}

## ‚úÖ Basic way
import json

data = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}

print(json.dumps(data))


Output:

{"name": "Alice", "age": 25, "city": "New York"}

‚úÖ Pretty-printed (formatted) JSON (recommended)
import json

print(json.dumps(data, indent=4))


Output:

{
    "name": "Alice",
    "age": 25,
    "city": "New York"
}

‚úÖ Sort keys alphabetically
print(json.dumps(data, indent=4, sort_keys=True))

üîë Important Notes

json.dumps() ‚Üí converts dictionary to JSON string

Dictionary keys must be strings

Python objects like set are not JSON serializable

Example error case:

data = {"numbers": {1, 2, 3}}  # set ‚ùå

Summary
Task	Method
Simple JSON	json.dumps(dict)
Pretty JSON	json.dumps(dict, indent=4)
Sorted keys	sort_keys=True
