## Generators

```python
## way to create iterator use a yield keyword to produce a series lazily 

def square(n):
    for i in range(3):
        yield i**2    ## save and return the variable also 
        
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a=square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

## Reading large files 

```python
## practical example :- Reading the large files 

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file :
            yield line
```

```python
file_path='large.txt'

for line in read_large_file(file_path):
    print(line.strip())
```

```python
## closures function :- function inside the function 

def main_welcome():
    msg="Welcome"   ## accessable inside the sub method also 
    
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn this concept clearly")
    
    return sub_welcome_method()
```

```python
main_welcome()
```

```python
def main_welcome(msg):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn this concept clearly")
    
    return sub_welcome_method()
```

```python
main_welcome("Welcome")
```

```python
## passing a functioon 

def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func("Welcome to the tutorial")
        print("Please learn this concept clearly")
    
    return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
## Decorators 

def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn this concept clearly")
    
    return sub_welcome_method()
```

```python
def course_introduction():
    print("This an advance python course")
    
course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
## use @ to create a decorator 
@main_welcome   ## the function below will be sent to the main_welcome and gets executed 
def course_introduction():
    print("This an advance python course")

```

```python
def my_decorator(func):
    def wrapper():
        print("Welcome to the advance python course")
        func()
        print("Please learn this concept clearly")
    return wrapper()
```

```python
@my_decorator
def say_hello():
    print("Hello1")
```

```python
def repeat(n):
    def my_decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return my_decorator
```

```python
@repeat(3)
def say_hello():
    print("hello!")
    
```

```python
say_hello()
```
