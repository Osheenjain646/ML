```python
## read a file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## read line by line

with open('example.txt','r') as file:
    for line in file:        ## take each line 
        print(line.strip())  ## .strip() removes the newline character(\n)
```

```python
## writing a file (overwriting)

with open('example.txt','w') as file:   # example.txt will be getting overwritten 
    file.write('Hello world1\n')
    file.write('this is a new line.')
```

```python
## without overwriting  
## Appending at the end of the file to protect from overwritten 

with open('example.txt','a') as file:
    file.write("\nAppend operation taking place!\n")
```

```python
## for writing the multiple or list of lines 

lines=['first line. \n' , 'second line. \n' , 'third line. \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary files (.bin file)

data=b'\x00\x01\x02\x03\x04'
with open('a.bin','wb') as file:
    file.write(data)
```

```python
## reading 

with open('a.bin','rb') as file:
    file.read()
```

```python
## writing and then reading a file

with open('example.txt','w+') as file:
    file.write("hello world\n")
    file.write("This is a new line.\n")
    
    ## move the file cursor to the begining
    
    file.seek(0)  ## take the cursor to the starting of the file
    
    ## read the content of the file
    
    content=file.read()
    print(content)
```

```python
## data serialization

import json
data={'name':'osheen','age':45}
## convert to json string
json_str=json.dumps(data)
print(type(json_str),json_str)

## convert back to dict fromjson
parsed_data=json.loads(json_str)
print(parsed_data,type(parsed_data))
```

```python
## csv file

import csv 
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['osheen',25])
    
with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row) 
```
