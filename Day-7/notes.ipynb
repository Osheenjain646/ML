```python
## try except else finally 

try:
    num=int(input("Enter the number: "))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter the denominator greater than 0")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")    ## will be executed if no exception occur
finally:
    print("Execution complete")   ## this will executed in the end always wheather a error is occur or not 

## use to do the final task that has to be done at the end always 
```

```python
## modelling a Bank Account 

## Define a class forr Bank Account 

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited.\n New Balance is {self.balance}")
        
    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient Balance")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn.\nThe new Balance is {self.balance}")
            
    def get_balance(self):
        return self.balance
    

## create an account 
account=BankAccount("Osheen",200000000000)
print(account.get_balance())

```

## Inheritance in python

```python
## Inheritence 
## parent class

class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
        
    def drive(self):
        print(f"The person will drive the {self.enginetype} engine car.")
```

```python
maclaren=Car(2,2,"V8")
buggati=Car(2,2,"v8")
buggati.drive()
```

```python
class Lamboghini(Car):  ## child(parent)
    def __init__(self, windows, doors, enginetype,speed):
        super().__init__(windows, doors, enginetype)    ## using the super to call parent and call init method from it 
        self.speed=speed
        
    def speedis(self):
        print(f"The top speed of lamborghini is {self.speed}.")
        

car1=Lamboghini(2,2,"v8",300)
car1.speedis()
```

```python
## Multiple Inheritence 
## when a class inherits frommore than one base class 
## base class 1
class Animal:
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print("subclass must implement this method")
        
## base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner
        
## derived class 
class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
        
    def speak(self):
        return f"{self.name} says woof"
    

## create an object 
dog3=Dog("Bruno","Osheen")
print(dog3.speak())
print(f"Owner: {dog3.owner}")
```

## Polymorphism 

```python
## Method Overriding 

## Base class 

class Animal:
    def speak(self):
        return "Sound of the animal"
    
## derived class 
class Dog(Animal):
    def speak(self):
        return "Woof"
    
## derived class 
class Cat(Animal):
    def speak(self):
        return "meow"
    

## function that demonstrate polymorphism

def animal_speak(animal):
    print(animal.speak())

dog=Dog()
cat=Cat()
print(dog.speak())
print(cat.speak())
animal_speak(dog)
print(animal_speak(cat))
```

```python
## polymorphism with function and methods 

## base class 

class Shape:
    def area(self):
        return "The area of the figure."
    
## derived class 1 
class Rectangle(Shape):
    def __init__(self,height,width):
        self.width=width
        self.height=height
        
    def area(self):
        return self.width*self.height
    

## derived class 2 
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
        
    def area(self):
        return 3.14*self.radius*self.radius
    
##  function that demonstrate polymorphism

def print_area(shape):
    print(f"The area is {shape.area()}")
    
rectangle=Rectangle(4,5)
circle=Circle(3)

print_area(circle)
print_area(rectangle)
```

```python
## polymorphism with abstract base class 

from abc import ABC,abstractmethod ## decorater 

## define an abstract class 

class Vehicle(ABC):
    @abstractmethod
    
    def start_engine(self):  ## abstract function must be an empty function 
        pass
    
## derived class 

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
## derived class 2

class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"
    
## function 

def start_engine(vehicle):
    print(f"The {vehicle.start_engine()}")
    
car1=Car()
motor=Motorcycle()

start_engine(car1)
start_engine(motor)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self._age=age
        self._name=name   ## protected variables now using __ cannot access otside the class but with derived class 
        self.gender=gender
    
class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
        
employee=Employee("osheen",45,"male")
print(employee._name)
```

```python
## Encapsulation with getter and setter 

class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age  ## private variable or access modifier 
        
    ## getter method for name 
    def get_name(self):
        return self.__name
    
    def set_name(self,name):
        self.__name=name
        
    ## setter method for age 
    
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("age can't be negative")
    
    def get_age(self):
        return self.__age
            

person=Person("osheen ",44)

## access and modify private variables using setter and getter

print(person.get_name())
print(person.get_age())

person.set_name("Anurag Raj")
person.set_age(90)

print(person.get_age())
print(person.get_name())

    
```

## Abstraction

```python
## to show only necessary features without showing the complex implementation , helps in reducing the programming complexity

## use abc for abstraction

from abc import ABC,abstractmethod

## Abstract base class 

class Vehicle(ABC):
    def derive(self):
        print("The vehicle is used for driving.")
        
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")
        
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.derive()
    
    
car=Car()
operate_vehicle(car)
```

## Operator Overloading

```python
## overwriting the operator functionality 

class Vector:
    def __init__(self,x,y):
        self.y=y
        self.x=x
    
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    def __mul__(self,other):
        return Vector(self.x*other.x,self.y*other.y)
    def __eq__(self,other):
        return Vector(self.x==other.x,self.y==other.y)
    def __repr__(self):
        return f"Vector ({self.x} , {self.y})"
    
## create objects of the Vector class
v1=Vector(2,3)
v2=Vector(4,5)

## it will take v2 as other 
## as v2 is second parameter 
print(v1+v2)
print(v1-v2)
print(v1*v2)
## use for truediv too
```
