## slicing in list

```python
numbers=[5,2,9,1,7,15,65,41,65,6]
print(numbers[2:5])
print(numbers[:4])
print(numbers[3:])
print(numbers[::2])
print(numbers[::-1])  # to reverse the list
```

```python
print(numbers[::-2])
```

```python
## iterating through a list
for number in numbers:
    print(number)
```

```python
## iterating through a list with index

for index,number in enumerate(numbers):
    print(index,number)
```

```python
## list comprehension
lst1=[]
for x in range(1,11):
    lst1.append(x*x)
print(lst1)
```

```python
## benefit of list comprehension
[x**2 for x in range(1,11)]
```

## list comprehension

syntax

[expression for item in iterable]

with condition [expression for item in iterable if condition]

```python
squares=[x**2 for x in range(1,11)]
print(squares)
```

```python
list2=[]
for x in range(1,21):
    if x%2==0:
        list2.append(x)
print(list2)
```

```python
even_numbers=[i for i in range(1,21) if i%2==0]
print(even_numbers)
```

## nested list comprehension



```python
## nested list comprehension

list2=[1,2,3,4,5]
list3=['a','b','c','d','e']

pair=[(i,j) for i in list2 for j in list3]
print(pair)
```

```python
## list comprehension with function calls

words=["hello","world","python","programming"]
lengths=[len(word) for word in words]
print(lengths)
```

## Tuples in python

```python
## these are immutable in nature in comparison to lists

tupl=()
print(tupl,type(tupl))
```

```python
lst=list() ## for empty list
print(lst,type(lst))
```

```python
numbers1=tuple([1,2,3,4,5])
numbers1
```

```python
(1,2,3,4,5,6)
list((1,2,3,4,5,6))
```

```python
mixed_tuple=("osheen",23,5.6,True)
mixed_tuple
```

```python
## accessing tuple elements

print(numbers1[0])
```

```python
## same slicing as lists for tuples 
print(numbers1[1:4])
```

```python
## tuple operations

concatenation_tuple=numbers1+mixed_tuple ## concatenation
print(concatenation_tuple)
```

```python
mixed_tuple*3 ## repetition
```

```python
numbers1*3 ## like lists, tuples also support repetition operation
```

```python
## immutability of tuples
# numbers1[0]=10  ## this will raise an error as tuples are immutable
tuple1=(1,2,3,1,1,5,41,15)
```

```python
## for changing a tuple, we can convert it to list, modify and convert back to tuple
list_from_tuple=list(tuple1)
list_from_tuple[0]=10
modified_tuple=tuple(list_from_tuple)
modified_tuple
```

```python
## tuples methods
tuple1.count(1) ## counts the occurrences of 1 in the tuple
tuple1.index(3) ## finds the index of first occurrence of 3 in the tuple
```

```python
## tuple unpacking and packing

packed_tuple="osheen",23,5.6
print(packed_tuple) ## packing as tuple
```

```python
## unpacking
## unpacking the values into variables as indexed by position
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with * 

number5=(1,2,3,4,5,6,7,8,9)
first,*second,rest=number5
print(first)
print(second)
print(rest)
```

```python
## nested tuples

nested_tuple=([1,2,3],[4,5,6],[7,8,9])
print(nested_tuple)
print(nested_tuple[0])
print(nested_tuple[1][2])
print(nested_tuple[2][1])
print(nested_tuple[0][0:3])

## tuple inside a tuple tuple inside a list and list inside a tuple and list inside a list

```

```python
## iterating through a nested tuple

for item in nested_tuple:
    for value in item:
        print(value , end=" ")
    print()
```
