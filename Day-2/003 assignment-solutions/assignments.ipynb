```python
## Assignment 12
numberfac=int(input("Enter the number whose factorail is required: "))
original_number=numberfac
factorial=1

while numberfac>=2:
    factorial*=numberfac
    numberfac-=1

print(f"The factorial of {original_number} is {factorial}")
```

```python
## Assignment 13

digitnumber=int(input("Enter the number whose digits sum is required: "))

digitsum=0
while digitnumber>0:
    digitsum+=digitnumber%10
    digitnumber=digitnumber//10

print(digitsum)
```

```python
## Assignment 14

primenumber=int(input("Enter the number too check whether it is prime or not: "))

flag=0
for i in range(1,primenumber+1):
    if primenumber%i==0:
        flag+=1
if flag==2:
    print(f"{primenumber} is prime.")
else:
    print(f"{primenumber} is not prime.")

```
