# Conditional Statement 

```python
## if statemnet

age=20
if age>=18:
    print("You are an adult.")
```

```python
age>=18 
```

```python
## else
## executes only if the if condition is false

age=16

if age>=18:
    print("You are an adult.")
else:
    print("You are an minor.")

```

```python
## elif statement
## allow you to check multiple conditions and stands for "else if"

age=20
if age<13:
    print("You are a child.")
elif age<18:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
## Nested conditional statements

## you can place one or more if statements inside another if statement

number = int(input("Enter a number: "))

if number > 0:
    print("The number is positive.")
    if number % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is negative or zero")

```

```python
## Practical Examples 

## Determining Leap Years

year = int(input("Enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, " is a leap year.")
        else:
            print(year, " is not a leap year.")
    else:
        print(year, " is a leap year.")
else:
    print(year , " is not a leap year.")


```

```python
##Assignment
## simple calculator

num1 = float(input("Enter first number: "))
operation = input("Enter operation (+, -, *, /): ")
num2=float(input("Enter the second number: "))

if operation=='+':
    result=num1+num2
elif operation=='-':
    result=num1-num2
elif operation=='*':
    result=num1*num2
elif operation=='/':
    if num2!=0:
        result=num1/num2
    else:
        result="Can't divide by zero"
else:
    result="Invalid Operation"

print(result)
```

```python
## Determine the ticket price based on the age and whether the person is a student .
## Ticket pricing is based on age and student status

age=int(input("Enter your age: "))
is_student=input("Are you a Student? (yes/no): ").lower()

if age<5:
    price="free"
elif age<=12:
    price="$10"
elif age<=17:
    if is_student =='yes':
        price="$12"
    else:
        price="$15"
elif age<=64:
    if is_student=='yes':
        price="$18"
    else:
        price="$25"
else:
    price="$20"

print("Ticket Price: ", price)
```

```python
range(5)
```

```python
## for loop

for i in range(5):
    print(i)

```

```python
for i in range(1,6):
    print(i)
    
```

```python
range(1,10,1)
## range(Start,end,Step)
```

```python
for i in range(1,10,1):
    print(i)

for i in range(1,10,2):
    print(i)

```

```python
for i in range(10,0,-1):
    print(i)
```

```python
## string

str1="Osheen Jain"

for i in str1:
    print(i)

```

```python
## while loop

## execute until a condition is true

count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
## loop control statemnets

## break

## the break exits the loop prematurely

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue

## skip the current iteration work in the loop 

for i in range(10):
    if i%2==0:
        continue
    print(i)

```

```python
## pass
## it is a null statemnet it does nothing

for i in range(10):
    if i==3:
        print("i is ",i)
        pass
    print(i)

```

```python
## nested loops
## loop inside a loop


for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
        
```

```python
##Example
## calculate the sum of N natural numbers 

## using while loop

n=10
sum=0
while n>0:
    sum=sum+n
    n=n-1

print("Sum of first 10 natural numbers: ",sum)
```

```python
n=10
sum=0
for i in range(n+1):
    sum=sum+i
    
print(sum)
```

```python
## Example - Prime number between 1 and 100

for i in range(1,101):
    if i>1:
        for j in range(2,i):
            if i%j==0:
                break
        else:
            print(i)

        
```
