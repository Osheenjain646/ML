## Data Manipulation and Analysis with Pandas 

```python
import pandas as pd 
```

```python
df = pd.read_csv('Students_data.csv')  ## use to read the csv file and create a data frame of it 
print(df.head(5))   ## give the first 5 rows of the data frame 
print(df.tail(5))   ## give the last 5 rows of the data frame 
```

```python
df.describe()   ## tells out the statistical data only for the numerical value only not for the categorical data 
```

```python
df.dtypes   ## provide the data types of the columns of the data frame 
```

```python
## Handling the missing values 
df.isnull()   ## give the boolean value of the data frame with true for null value and false for non null value
```

```python
df.isnull().any(axis=0)  ##tell if there is any null value in the column or not 
```

```python
df.isnull().any()  ## give the boolean value of the data frame with true for null value and false for non null value for each column
```

```python
df.isnull().sum()  ## give the count of null value in each column 
```

```python
df_fillna=df.fillna(0)  ## fill the null value with 0
print(df_fillna)
```

```python
## filling the missing value with mean of the column 
df['Algebra_fillna']=df['Algebra'].fillna(df['Algebra'].mean()) ## fill the null value of the sales column with the mean of the sales column in a new column named as sales_fillna or algebra_fillna for algebra column 
df
```

```python
df.dtypes
```

```python
## Remaing Columns 
df=df.rename(columns={'ID':'Student_Id'})
df
```

```python
## Change the data types 
df['new_Algebra']=df['Algebra'].fillna(df['Algebra'].mean()).astype(int)  ## fill the null value of the Algebra column with the mean of the algebra column and change the data type to int in a new column named as new_Algebra 
df
```

```python
df['new_race']=df['race'].apply(lambda x:x*2)   ## put function in the apply 
df
```

## Merging and Joining the Data Frame 

```python
## creating the dataframe 

df1=pd.DataFrame({'Key':['A' , 'B' , 'C'] , 'values':[1,2,3]})
df2=pd.DataFrame({'Key':['A' , 'B' , 'D'] , 'values':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge DataFrame on the 'Key columns' 
df1=pd.merge(df1,df2,on='Key' , how="inner")  ## how can be inner,outepd.merge(df1,df2,on='Key' , how="outer")
pd.merge(df1,df2,on='Key' , how="left")
pd.merge(df1,df2,on='Key' , how="right")
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot 
plt.plot(x,y)  ## create a line plot with x and y values 
plt.xlabel('X-axis')  ## label for x axis
plt.ylabel('Y-axis')  ## label for y axis
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a customized line plot 
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)  ## create a line plot with x and y values with red color, dashed line style and circle marker
plt.grid(True)
plt.xlabel('X-axis')  ## label for x axis
plt.ylabel('Y-axis')  ## label for y axis
plt.title('Customized line plot')
plt.show()
```

```python
## multiple plots 
## sample plots 
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5)) ## set the figure size for the plot
plt.subplot(2,2,1)  ## create a subplot with 1 row and 2 columns 
plt.plot(x,y1,color='green')
plt.title("plot 1 ")

plt.subplot(2,2,2)  
plt.plot(x,y2,color='blue')
plt.title("plot 2 ")

plt.subplot(2,2,3)  
plt.plot(y1,y2,color='red')
plt.title("plot 3 ")

plt.subplot(2,2,4)  
plt.plot(y1,x,color='orange')
plt.title("plot 4 ")
```

```python
## bar plot 

categories=['A','B','C','D','E']
values=[30,50,10,40,20]

## create a bar plot 
plt.bar(categories,values,color='purple')  ## create a bar plot with categories on x axis and values on y axis with purple color
plt.xlabel('Categories')  ## label for x axis
plt.ylabel('Values')  ## label for y axis
plt.title('Bar Plot')  ## title for the plot
plt.show()
```

```python
import seaborn as sns 
```

```python
## Basic plotting with seaborn 
tips=sns.load_dataset('tips')
tips
```

```python
## create a scatter plot 
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips,color='red') ## create a scatter plot with total_bill on x axis and tip on y axis with red color
plt.title('Scatter Plot of Total Bill vs Tip')  ## title for the plot
plt.show()
```

```python
## Line plot with seaborn

sns.lineplot(x='size',y='total_bill',data=tips,color="green")
plt.show()
```

```python
## categorical plot with seaborn
## bar plot 

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot with seaborn')  ## title for the plot
plt.show()
```

```python
## box plot 

sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
## voilin plot 

sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
## histogram 

sns.histplot(tips['total_bill'],bins=10,kde=True)  ## create a histogram with total_bill column and 10 bins with kernel density estimation
plt.title('Histogram of Total Bill') 
plt.show()
```

```python
## KDE plot 

sns.kdeplot(tips['total_bill'],fill=True) ## shade is depriciated
plt.show()
```

```python
## pairplot 

sns.pairplot(tips)
plt.show()
```

```python
## heatmap

corr=tips[['total_bill','tip','size']].corr()  ## calculate the correlation matrix of the tips dataset
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')  ## create a heatmap with correlation matrix with annotations and coolwarm color map
plt.title('Correlation Heatmap')  ## title for the plot
plt.show()
```
