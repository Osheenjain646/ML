## Python Logging



```python
import logging
```

```python
## first configure the basic logging setting 

logging.basicConfig(level=logging.DEBUG)

## log meassages with different severity levels

logging.debug("This is a debug meassage")
logging.info("This is info message")
logging.warning("This is warning message")
logging.error("This is a error message")
logging.critical("This is a critical message")
```

## 


```python
## configuring logging 

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True   ## to do the time and date format 
)
## log meassages with different severity levels

logging.debug("This is a debug meassage")
logging.info("This is info message")
logging.warning("This is warning message")
logging.error("This is a error message")
logging.critical("This is a critical message")
```

## Logging with Multiple Loggers

```python
import logging 
```

```python
## create a logger for module1 
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## create a logger for module2 

logger2=logging.getLogger("module2")
logger2.setLevel(logging.DEBUG)

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True   
)
```

```python
## log messages with different messages 

logger1.debug("This is a debug message for logger 1")
logger2.warning("This is a warning message for logger 2")
logger2.error("This ia a error message for logger 2")

```
